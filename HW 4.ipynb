```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
df = pd.read_csv('kc-house-data.csv')
df
```

*1.2. Изучите стоимости недвижимости*

```python
plt.figure(figsize=(20, 8))
sns.histplot(df['price'])
plt.ticklabel_format(style='plain')
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество');
```

Вывод по стоимости:
Стоимость основного количества домов ~ 1 млн.

*1.3. Изучите распределение квадратуры жилой*

```python
plt.figure(figsize=(20, 6))
sns.histplot(df['sqft_living'])
plt.ticklabel_format(style='plain')
plt.title('Распределение жилой площади')
plt.xlabel('Площадь')
plt.ylabel('Количество');
```

Вывод по площади: Максимальное количество недвижимости продается с площатью примерно от 1700 до 2200 кв.м.

*1.4. Изучите распределение года постройки*

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['yr_built'])
plt.ticklabel_format(style='plain')
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество');
```

*Вывод по распределению года постройки:*

Наблюдается кризис в строительстве в 1930-х годах, а также спад в середине 2000-х годов.
Наибольшее число построеных домов наблюдается в конце 1970-х годов и начале 2000-х годов.

**Условие 3: 3 задача***

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Корреляция');
```

Матрица корреляции показывает существенную связь цены дома с:
- количеством ванных (bathrooms)
- жилой площадью (sqft_living)
- классом дома (grade)
- площадью этажей (sqft_above)
-  Количетво этажей (floors)

 *Графики стоимости домов в зависимости от параметров:*

```python
sns.boxplot(x=df['bathrooms'], y=df['price'], showfliers=False)
plt.title('Зависимость цены дома от количества ванных')
plt.xticks(rotation=50);
```

```python
sns.scatterplot(x=df['sqft_living'], y=df['price'])
plt.title('Зависимость цены дома от жилой площади');
```

```python
sns.boxplot(x=df['grade'], y=df['price'], showfliers=False)
#sns.boxplot(x=df['grade'], y=df['price'], showfliers=True)
plt.title('Зависимость цены дома от класса')
plt.xticks(rotation=50);
```

```python
sns.scatterplot(x=df['sqft_above'], y=df['price'])
plt.title('Зависимость цены дома от площади этажей');
```

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('Стоимость')
plt.ylabel('Этажность')
plt.title('Зависимость цены дома от этажности');
```

**Основные выводы:**
1. С увеличением количества ванных комнат увеличивается стоимость дома.
2. Стоимость дома увеличивается с ростом его площади.
3. У домов с котегориями от 7 по 13 стоимость выше.
4. Стоимость ростет с увеличением площади этажей.
5. Наиболее дорогие дома имеют этажность 2, а самую низкую стоимость дома этажностью 3,5.
